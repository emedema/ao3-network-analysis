# Tag Network Analysis

Let's look at the network generated by the tags on AO3. In order to use the data with Gephi to generate a social network graph, we need to put the data into the nodes/edges format.

```python
#imports
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
import warnings; warnings.simplefilter('ignore')
import json
import pandas as pd
```

```python
#read in data
tag_df = pd.read_json('../ao3bot/tag_network.json')
tag_df.head()
```

```python
nodes = []
nodes_unformatted = []
tags = tag_df["tag"].tolist()
count = 1
for tag in tags:
    l = tag_df.loc[tag_df["tag"] == tag]["freeforms"]
    if len(l) > 0:
        for i in l:
            for j in i:
                if j not in nodes_unformatted:
                    if "*s*" in j:
                        j = j.replace("*s*", "/")
                    nodes.append([count, j])
                    nodes_unformatted.append(j)
                    count += 1

print(len(nodes))
# nodes_df = pd.DataFrame(nodes, columns=["Id","Label"])
# nodes_df.head()
```

```python
nodes_df = pd.DataFrame(nodes, columns=["Id","Label"])
nodes_df.head()
nodes_df.to_csv("nodes_and_edges/tag_nodes.csv", index = False)
```

```python
edges = []
tags = tag_df["tag"].tolist()
for tag in tags:
    l = tag_df.loc[tag_df["tag"] == tag]["freeforms"]
    if len(l) > 0:
        for i in l:
            for j in i:
                if "*s*" in tag:
                    tag = tag.replace("*s*", "/")
                target = nodes_df.loc[nodes_df.Label == j,'Id'].tolist()[0]
                src = nodes_df.loc[nodes_df.Label == tag,'Id'].tolist()[0]
                edges.append([src, target, "Undirected"])

print(len(edges))
```

```python
edges_df = pd.DataFrame(edges, columns = ["Source", "Target", "Type"])
edges_df
```

```python
edges_df.to_csv("nodes_and_edges/tags_edges.csv", index = False)
```
